```python
##범죄별 그래프

import csv
import matplotlib.pyplot as plt
f = open('범죄.csv')
plt.rc('font', family = 'AppleGothic')
data = csv.reader(f)
next(data)
data = list(data)
for i in range(0, 38):
    a = data[i]
    pivot=[]
    for j in range(4, 11):
        b=int(a[j])
        pivot.append(b)
    plt.plot(pivot)
    plt.title(a[1])
    plt.style.use('ggplot')
    plt.show()
```

```python
import csv
f = open('범죄.csv')

data = csv.reader(f)
next(data)
data = list(data)
crime = []

name = input('범죄 종류 입력')

for row in data :
    if name in row[1] :
        for i in range(4, 11) : 
            crime.append(int(row[i])/int(row[3]))
        break
rate = 1

for row in data :
    s = 0
    for i in range(4, 11) :
        row[i] = int(row[i])/int(row[3])
        tmp = (row[i] - crime[i-4]) ** 2
        s = s + tmp
    if s < rate and (name not in row[1]) :
        result = []
        for i in range(4, 11) :
            result.append(row[i])
        rate = s            
        result_name = row[1]
            
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.plot(crime, label = name)
plt.plot(result, label = result_name)
plt.rc('font', family = 'AppleGothic')
plt.legend()
plt.show()
```
